## 2. Model Implementation

**Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.**

**Divide the dataset into training and test sets.**

**Train your model on the training set and then use it to make predictions on the test set.**


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

model = Sequential()

# Input Layer + Hidden Layer
model.add(Dense(32, input_dim=60, activation='relu'))

# Second Hidden Layer
model.add(Dense(16, activation='relu'))

# Output Layer
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile Model
model.compile(optimizer=Adam(learning_rate=0.001),
             loss='binary_crossentropy',
             metrics=['accuracy']
             )
```

```python
model.summary()
```

```python
# Train Model

history = model.fit(X_train, y_train, 
                   validation_split=0.2,
                   epochs=100,
                   batch_size = 16,
                   verbose=1
                   )
```

```python
# Evaluation on test data

loss, accuracy = model.evaluate(X_test, y_test)
print('Test Accuracy: ', accuracy)
```

```python
# Make prediction

y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

print('Predictions: \n', y_pred[:5])
```

The dataset was divided into training and testing sets using an 70вҖ“30 split with stratification to maintain class balance. A Sequential neural network model was implemented using TensorFlow/Keras.

The architecture consisted of:

* One input layer with 60 neurons (corresponding to the 60 sonar features)

* Two hidden layers with ReLU activation functions to capture nonlinear patterns

* One output layer with a sigmoid activation function for binary classification

The model was compiled using the Adam optimizer and binary cross-entropy loss function. After training the model on the training data for multiple epochs, predictions were generated on the test dataset.

## 3. Hyperparameter Tuning

**Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance**

**Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.**


```python
# Function so that GridSearch can try different combinations.

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

def create_model(neurons=16, activation='relu', learning_rate=0.001):
    
    models = Sequential()
    
    # Hidden Layer 1
    models.add(Dense(neurons, input_dim=60, activation=activation))
    
    # Hidden Layer 2
    models.add(Dense(neurons//2, activation=activation))
    
    # Output Layer
    models.add(Dense(1, activation='sigmoid'))
    
    optimizer = Adam(learning_rate=learning_rate)
    
    models.compile(optimizer=optimizer,
                 loss='binary_crossentropy',
                 metrics=['accuracy']
                 )
    
    return models
```

```python
# Wraping Keras deep learning model so that it can be treated as a standard Scikit-Learn estimator.

from scikeras.wrappers import KerasClassifier

models = KerasClassifier(model=create_model, verbose=0)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'model__neurons':[16,32,64],
    'model__activation':['relu', 'tanh'],
    'model__learning_rate':[0.001, 0.01],
    'batch_size':[8,16],
    'epochs':[50, 100]
}

grid = GridSearchCV(estimator=models,
                   param_grid=param_grid,
                   cv=5,
                   n_jobs=-1
                   )
```

```python
# Fit Grid Search

%time grid_result = grid.fit(X_train, y_train)
```

```python
print("Best Accuracy: ", grid_result.best_score_)
print("Best Parameters: ", grid_result.best_params_)
```

```python
best_model = grid_result.best_estimator_

# Evaluate on test data
test_accuracy = best_model.score(X_test, y_test)

print("Test Accuracy after Tuning:", test_accuracy)
```

A Grid Search with 5-fold cross-validation was used to systematically evaluate different combinations of hyperparameters, including:

* Number of neurons in hidden layers

* Activation function

* Learning rate

* Batch size

* Number of epochs

After evaluating multiple combinations, the best performance was achieved with the following configuration:

* Neurons: 32

* Activation Function: ReLU

* Learning Rate: 0.001

* Batch Size: 8

* Epochs: 50

This configuration resulted in a best cross-validation accuracy of 84.83%, indicating strong learning capability during validation. When evaluated on the unseen test dataset, the tuned model achieved a test accuracy of 79.37%.

## 4. Evaluation

**Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.**


```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predictions from default model
y_pred_default = (model.predict(X_test) > 0.5).astype(int)

# Predictions from tuned model
y_pred_tuned = best_model.predict(X_test)

print("Default Model Evaluation")
print("Accuracy:", accuracy_score(y_test, y_pred_default))
print("\nClassification Report:\n", classification_report(y_test, y_pred_default))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_default))

print("\nTuned Model Evaluation")
print("Accuracy:", accuracy_score(y_test, y_pred_tuned))
print("\nClassification Report:\n", classification_report(y_test, y_pred_tuned))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tuned))
```
